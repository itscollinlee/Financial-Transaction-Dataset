# Create Finalized consumption table with all intermediary silver tables

```python
%sql
CREATE OR REPLACE TABLE financial_transactions_dataset.default.silver_transactions_id
AS 
(SELECT 
 A.id
,A.transaction_date
,A.transaction_timestamp
,A.client_id
,D.current_age
,D.retirement_age
,D.birth_year
,D.birth_month
,D.gender
,D.address
,D.latitude
,D.longitude
,D.per_capita_income
,D.yearly_income
,D.total_debt
,D.credit_score
,D.num_credit_cards
,A.card_id
,E.account_open_date
,E.account_open_year
,E.account_open_month
,E.expiry_date
,E.expiry_year
,E.expiry_month
,E.year_pin_last_changed
,E.card_brand
,E.card_type
,E.card_number
,E.cvv
,E.has_chip
,E.num_cards_issued
,E.credit_limit
,E.card_on_dark_web
,A.amount
,A.use_chip
,A.merchant_id
,A.merchant_city
,A.merchant_state
,A.zip_code
,A.mcc_code
,B.mcc_code_description
,C.fraud_flag
,A.errors
FROM financial_transactions_dataset.default.silver_transactions_data A
LEFT JOIN financial_transactions_dataset.default.silver_mcc_codes_data B 
ON A.mcc_code = B.mcc_code
LEFT JOIN financial_transactions_dataset.default.silver_train_fraud_labels C 
ON A.id = C.id 
LEFT JOIN financial_transactions_dataset.default.silver_users_data D 
ON A.client_id = D.client_id 
LEFT JOIN financial_transactions_dataset.default.silver_cards_data E 
ON A.card_id = E.card_id
AND A.client_id = E.client_id
)
```
